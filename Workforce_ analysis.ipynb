```python
# First let us import the neccessary lebraries and data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Here we have 2020 and 2021 datasets and we are going to upload and explore them monthly basis then we compare the results.
# First we will go through 2020 datasets then 2021.
data = pd.read_csv('During_covid.csv')
data.count()
```

```python
data.info()
```

```python
Online_order = data['Top Trending Jobs'].value_counts()
plt.figure(figsize = (15,10))
Online_order.plot(kind = 'barh', color = 'red')
plt.ylabel("Frequency")
plt.show()
```

```python
Online_order = data['Top Trending Skills'].value_counts()
plt.figure(figsize = (18,10))
Online_order.plot(kind = 'barh', color = 'blue')
plt.ylabel("Frequency")
plt.show()
```

```python
Online_order = data['Top Trending Employers'].value_counts()
plt.figure(figsize = (10,5))
Online_order.plot(kind = 'bar', color = 'purple')
plt.ylabel("Frequency")
plt.xticks(rotation = 75)
plt.show()
```
