### EDA (Exploratory Data Analysis)

```python
# First let us import the neccessary lebraries and data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Uploading the dataset
data = pd.read_csv('During_covid.csv')
data.count()
```

```python
data.info()
```

## Date Gathered from Linkden.com

### Top Trending Jobs

```python
Online_order = data['Top Trending Jobs'].value_counts()
plt.figure(figsize = (15,10))
Online_order.plot(kind = 'barh', color = 'red')
plt.ylabel("Frequency")
plt.show()
```

### Top Trending Skills

```python
Online_order = data['Top Trending Skills'].value_counts()
plt.figure(figsize = (18,10))
Online_order.plot(kind = 'barh', color = 'blue')
plt.ylabel("Frequency")
plt.show()
```

### Top Trending Employers

```python
Online_order = data['Top Trending Employers'].value_counts()
plt.figure(figsize = (10,5))
Online_order.plot(kind = 'bar', color = 'purple')
plt.ylabel("Frequency")
plt.xticks(rotation = 75)
plt.show()
```
